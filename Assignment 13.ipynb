```python
import sqlalchemy
from sqlalchemy import create_engine
import numpy as np
import pandas as pd
import sqlite3
```

```python
#import the data
AdultdataDF = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data')
AdultdataDF.head()
```

```python
#assign column headings
AdultdataDF.columns =['Age','Workclass','fnlwgt','Education','EducationNum','MaritalStatus','Occupation','Relationship','Race','Sex','CapitalGain','CapitalLoss','HoursPerWeek','NativeCountry','Salary']
AdultdataDF.head()
```

```python
#1 start the engine and connect to the dataframe via cursor
engine = create_engine('sqlite:///adultdata2', echo=False)
AdultdataDF.to_sql('adultdata2',engine,if_exists='replace')
connection = sqlite3.connect("adultdata2")
cursor= connection.cursor()
```

```python
#2 write two update queries
cursor.execute('UPDATE adultdata2 SET CapitalGain = 1 WHERE CapitalGain =0')
cursor.execute("UPDATE adultdata2 SET HoursPerWeek = 39 WHERE HoursPerWeek = 40")
print(pd.read_sql_query("SELECT * from adultdata2", connection).head())
```

```python
#3 two delete queries
cursor.execute('DELETE from adultdata2 WHERE EducationNum = 13')
cursor.execute('DELETE from adultdata2 WHERE HoursPerWeek = 13')
print(pd.read_sql_query("SELECT * from adultdata2", connection).head())
```

```python
#4 two filter queries
print(pd.read_sql_query('SELECT * FROM adultdata2 WHERE Age > 40',connection).head())
print(pd.read_sql_query('SELECT * FROM adultdata2 WHERE EducationNum < 14',connection).head())
```

```python
#two function queries
print(pd.read_sql_query('SELECT MAX(HoursPerWeek) FROM adultdata2',connection).head())
print(pd.read_sql_query('SELECT AVG(HoursPerWeek) FROM adultdata2',connection).head())
```
